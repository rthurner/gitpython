# Yahoo export formatting

```python
import pandas as pd
```

```python
Quotes=pd.read_csv("data/Yahoq.csv")
```

```python
Quotes.head()
```

```python
Qgrop = Quotes[['Symbol','Current Price','Purchase Price','Quantity']]
```

```python
Qgrop=Qgrop[Qgrop['Quantity'].notna()]
```

```python
Qgrop['Market']=Qgrop['Current Price'] * Qgrop['Quantity']
```

```python
Qgrop['Cost']=Qgrop["Purchase Price"] * Qgrop["Quantity"]
```

```python
Dgroup=Qgrop.groupby('Symbol').agg({'Current Price':['mean','count'],'Quantity':['sum'],
                                   'Market':['sum'], 'Cost':['sum'],'Purchase Price':['mean']})
```

```python
Dgroup.columns = ['C_Price','Count','Quantity','Market','Cost','Price']
```

```python
Dgroup['Prof']=Dgroup["Market"] - Dgroup['Cost']
```

```python
Dgroup.loc['ZSUM']=Dgroup.sum()
```
